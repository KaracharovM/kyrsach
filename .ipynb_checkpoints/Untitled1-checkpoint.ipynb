4.Пробуем составить график прогноза курса юани относительно доллара.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
from math import sqrt

# Загрузка данных из файла CSV
currency = pd.read_csv("C:\\EXCHUS.csv")
currency['DATE'] = pd.to_datetime(currency['DATE'])
currency.set_index('DATE', inplace=True)

# Разделение данных на обучающую и тестовую выборки
train_size = int(len(currency) * 0.8)
train, test = currency[0:train_size], currency[train_size:]

# Обучение модели экспоненциального сглаживания
model = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=5)
model_fit = model.fit()

# Прогноз на тестовой выборке
predictions = model_fit.forecast(len(test))

# Визуализация прогноза и фактических значений
plt.plot(test.index, test.values, label='Фактический курс')
plt.plot(test.index, predictions, color='red', label='Прогноз курса')
plt.title('Прогноз курса юани с использованием экспоненциального сглаживания')
plt.xlabel('Дата')
plt.ylabel('Курс')
plt.legend()
plt.show()

# Оценка точности модели
rmse = sqrt(mean_squared_error(test, predictions))
print(f'RMSE: {rmse}')
```

Описание : Прогноз курса составленный на основе 8 лет методом экспоненциального сглаживания.
